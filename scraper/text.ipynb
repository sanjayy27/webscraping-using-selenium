```python
from selenium import webdriver
import time
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://www.rfpage.com/applications-5g-technology/')
```

```python
time.sleep(3)
content=driver.find_elements(By.TAG_NAME,"h2")
content1=driver.find_elements(By.TAG_NAME,"h3")
```

```python
content1
```

```python
uses=[]
def collector(x):
    for i in x:
        a=i.text
        uses.append(a)
```

```python
collector(content)
collector(content1)
```

```python
uses
```

```python
df=pd.DataFrame(uses,columns=['uses'])
df
```
